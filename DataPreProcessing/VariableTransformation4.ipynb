```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

0-1 Transformation

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
df["new_sex"] = lbe.fit_transform(df["sex"])
```

```python
df
```

1-Others(0) Transformation

```python
import numpy as np
df["new_day"] = np.where(df["day"].str.contains("Sun"), 1 ,0)
df.head(25)
```

Multi-Variable Transformation

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["day"])
```

WARNING! there is no diff between the observs of the day (day is a nominal categorical variable).

However if we transform the 'day' values to numerical, it looks like there is a difference between them. For example, 8 is greater than 0.

The solution is One-Hot Transformation

```python
df.head()
```

```python
import pandas as pd

df_one_hot = pd.get_dummies(df, columns=["sex"], prefix=["sex"])
df_one_hot.head()
```

sex_male and sex_female are actually same thing. It's called the "Dummy Variable Trap".

```python
pd.get_dummies(df, columns=["day"], prefix=["day"]).head()
```

Continously Variable to Categorical Variable

```python
df = df.drop(columns="new_sex")
df = df.drop(columns="new_day")
df
```

```python
dff = df.select_dtypes(include=["float64", "int64"])
dff
```

```python
from sklearn.preprocessing import KBinsDiscretizer
est = KBinsDiscretizer(n_bins=[3,2,2], encode="ordinal", strategy="quantile").fit(dff)
```

```python
est.transform(dff)[0:10]
```
