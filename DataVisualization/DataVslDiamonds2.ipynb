```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df['cut'].value_counts()
```

```python
df['color'].value_counts()
```

```python
df['clarity'].value_counts()
```

```python
#ordinal tanımlama
import pandas.api.types as ptypes
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(ptypes.CategoricalDtype(categories = cut_kategoriler, ordered = True))
```

```python
df.cut.head(1)
```

```python
(df['cut']
 .value_counts()
 .plot.barh()
 .set_title('Cut Değişkeninin Sınıf Frekansları'));
```

BARPLOT VE CATPLOT

```python
sns.barplot(x = 'cut', y = df.cut.index, data = df);
```

```python
sns.catplot(x = 'cut', y = 'price', data = df);
```

```python
sns.barplot(x = 'cut', y = 'price', hue = 'color', data = df);
#burada price değerlerinin ortalamasını aldık cut ve color'a göre
```

```python
#bunu da şöyle doğrulayalım
df.groupby(['cut','color'], observed=True)['price'].mean()
```

HISTOGRAM VE YOĞUNLUK

Sayısal değişkenleri görselleştirmek için kullanılır

```python
sns.histplot(df.price, bins = 50);
```

```python
sns.kdeplot(df.price, fill = True);
```

HISTOGRAM VE YOĞUNLUK ÇAPRAZLAMALAR

```python
(sns.FacetGrid(df,
               hue = 'cut',
               height = 5,
               xlim = (0,10000))
 .map(sns.kdeplot, 'price', fill = True)
 .add_legend()
);
```

```python
sns.catplot(x = 'cut', y = 'price', hue = 'color', kind = 'point', data = df);
```
